```python
import torch 
import torch.nn as nn
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import numpy as np 
import math 
import scipy.io as io
import PIL.Image as Image
import numpy as np
import os
from matplotlib import pyplot as plt
from scipy.ndimage import gaussian_filter 
import scipy
from matplotlib import cm as CM
import torch
import csv
%matplotlib inline

device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
print(device)
```

```python
actual_dir_train = "./FinalModelCounts/personActual_train.txt" 
YOLO_dir_train = "./FinalModelCounts/personYOLO_train.txt"

actual_dir_test = "./FinalModelCounts/personActual_test.txt" 
YOLO_dir_test = "./FinalModelCounts/personYOLO_test.txt"
```

## Actual vs Predicted, Train

```python
predicted_cnt_list = []
actual_cnt_list = []

with open(YOLO_dir_train,newline='') as file:
    reader = csv.reader(file)
    YOLO_dict = dict(reader)  # pull in each row as a key-value pair
with open(actual_dir_train,newline='') as file:
    reader = csv.reader(file)
    actual_dict = dict(reader)  # pull in each row as a key-value pair


for idx in range(len(actual_dict)):
  print("|", end="")
  yoloquery ="images_train/" + f"IMG_{idx+1}.jpg"
  actualquery ="images_train/" + f"IMG_{idx+1}.jpg"
  YOLO_count = int(YOLO_dict[yoloquery])
  actual_count = float(actual_dict[actualquery])

  predicted_cnt_list.append(YOLO_count)
  actual_cnt_list.append(actual_count)
```

```python
plt.scatter(actual_cnt_list, predicted_cnt_list)
x = np.linspace(0,1600,100)
y = x
plt.plot(x, y, '-r', label='y=x')

plt.title("Actual vs Predicted, Train")
# plt.xlabel(f"Actual Count\n\nAverage MAE: {sum(loss_list) / len(loss_list)}")
plt.xlabel(f"Actual Count")
plt.ylabel("Predicted Count")
plt.grid()
plt.show()
```

```python
plt.scatter(actual_cnt_list, predicted_cnt_list)
x = np.linspace(0,1600,100)
y = x
plt.plot(x, y, '-r', label='y=x')

plt.title("Actual vs Predicted, Train (lim=100)")
# plt.xlabel(f"Actual Count\n\nAverage MAE: {sum(loss_list) / len(loss_list)}")
plt.xlabel(f"Actual Count")
plt.ylabel("Predicted Count")
plt.grid()
plt.xlim([0, 100])
plt.ylim([0, 100])

plt.show()
```
